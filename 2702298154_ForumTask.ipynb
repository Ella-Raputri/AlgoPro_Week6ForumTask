```python
#Forum Task 1

def average(numbers):
    total = sum(numbers)
    total = float(total)
    return (total/len(numbers))

def get_average(student):
    homework = average(student["homework"])
    quizzes = average(student["quizzes"])
    tests = average(student["tests"])
    return ((10/100 * homework) + (30/100 * quizzes) +(60/100 * tests))

def get_letter_grade(score):
    score = float(score)

    if (score >= 90):
        return "A"
    elif (score >= 80):
        return "B"
    elif (score >= 70):
        return "C"
    elif (score >= 60):
        return "D"
    else:
        return "F"
    
def get_class_average(students):
    results = []
    for student in students:
        results.append(get_average(student))
    return (average(results))

#The student's dictionary
lloyd = {
 "name": "Lloyd",
 "homework": [90.0,97.0,75.0,92.0],
 "quizzes": [88.0,40.0,94.0],
 "tests": [75.0,90.0]
}
alice = {
 "name": "Alice",
 "homework": [100.0, 92.0, 98.0, 100.0],
 "quizzes": [82.0, 83.0, 91.0],
 "tests": [89.0, 97.0]
}
tyler = {
 "name": "Tyler",
 "homework": [0.0, 87.0, 75.0, 22.0],
 "quizzes": [0.0, 75.0, 78.0],
 "tests": [100.0, 100.0]
}

#Print the students dictionary's keys and values
students = [lloyd, alice, tyler]
for student in students:
    for key,value in student.items():
        print(f"{key.title()}:{value}")
    print()

#Test for the get_letter_grade function
print("The letter grade for Lloyd is",get_letter_grade(get_average(lloyd)))

#Print the result of get_class_average for all students
print("The class average is",get_class_average(students))

#Print the result of get_letter_grade for class average
print("The letter grade for the class average is",get_letter_grade(get_class_average(students)))
```

```python
#Forum Task 3

#Days of a week
week = ("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday")

#Prompt users to enter retail sales amount
list_revenue = []
for day in range(0, len(week)):
    revenue = input(f"Enter the sales for {week[day]}: ")
    while revenue.isdigit() == False:
        revenue = input(f"Input was invalid. Re-enter the sales for {week[day]}: ")
    
    revenue = int(revenue)
    list_revenue.append(revenue)

total_sales = sum(list_revenue)
minimum_sales = min(list_revenue)
maximum_sales = max(list_revenue)

print(f"The total sales is ${total_sales:,.2f}")
print(f"The minimum sale amount was: ${minimum_sales:,.2f}")
print(f"The maximum sale amount was: ${maximum_sales:,.2f}")
```
